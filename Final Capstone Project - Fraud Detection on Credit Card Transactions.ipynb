```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('creditcard.csv')
data
```

```python
data.describe()
```

```python
sns.distplot(data['Amount'])
```

The data in the AMOUNT column is right-skewed. The data in that column is composed majorly of small numbers. Large numbers are very few

```python
X=data.iloc[:,:30].values
y=data.iloc[:,30].values
```

### Splitting the data into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

### Applying feature scaling to the data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

### Applying K-Nearest Neighbors classifier

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
